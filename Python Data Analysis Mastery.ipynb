```python
root["joining_date"] = pd.to_datetime(root["joining_date"], dayfirst=True)
px.line(root.groupby(root["joining_date"].dt.year).size().reset_index(name="count"),
        x="joining_date", y="count", title="Employees Joined per Year")
```
